## 2.7 사용자 정의 함수
* 파이썬 내장 함수: type(), len(), ord() 등
* 사용자 정의 함수

```python
def find_max(A):
    max = A[0]
    for i in range(1, len(A)):
        if max < A[i]:
            max = A[i]
    return max
```

```python
data = [5,3,8,4,9,1,6,2,7]
print("max =", find_max(data))
```

```python
# 여러 개의 값을 반환
def find_min_max(A):
    min = A[0]
    max = A[0]
    for i in range(1, len(A)):
        if max < A[i]:
            max = A[i]
        if min > A[i]:
            min = A[i]
    return min, max
```

```python
data = [5,3,8,4,9,1,6,2,7]
x, y = find_min_max(data)
print("min, max =", x, ',', y)
```

```python
# 디폴트 인수
def sum_range(begin, end, step=1):      # 매개변수 step이 기본 값을 가짐
    sum = 0
    for n in range(begin, end, step):
        sum += n
    return sum
```

```python
print(sum_range(1, 10))
print(sum_range(1, 10, 2))
```

전역 변수

```python
def calc_perimeter(radius):
    # global perimeter
    print("파이 값:", pi)
    perimeter = 2*pi * radius

pi = 3.14159
perimeter = 0
calc_perimeter(10)
print("원둘레(r=10) =", perimeter)
```

pi        : 함수 내에서 값이 사용됨<br/>
perimeter : 함수 내에서 값을 변경 => 함수 내 global perimeter 선언문 필요 => global 선언문 없으면 함수 내 지역변수로 새로 생성됨

```python
def g_calc_perimeter(radius):
    global perimeter
    print("파이 값:", pi)
    perimeter = 2*pi * radius

pi = 3.14159
perimeter = 0
g_calc_perimeter(10)
print("원둘레(r=10) =", perimeter)
```

## 2.10 클래스
* 공통된 것들을 추상화 시킨것

```python
# 클래스 정의
class Car:  # 자동차와 관련된 클래스 정의 블록이 이어진다. 들여쓰기 주의
    def __init__(self, color, speed=0):     # 생성자 함수
        self.color = color                  # 데이터 멤버 color 정의 및 초기화
        self.speed = speed

    # 멤버함수
    def speedUp(self):
        self.speed += 10
    def speedDown(self):
        self.speed -= 10
```

```python
car1 = Car('black', 0)
car2 = Car('red', 120)
car3 = Car('green')

car2.color
print(car3.speed)
car3.speedUp()
print(car3.speed)
```

## 2.11 연신자 중복

```python
class Car :
    def __init__(self, color, speed = 0) :	
        self.color = color		            
        self.speed = speed		            
    def speedUp(self) : self.speed += 10	
    def speedDown(self) : self.speed -= 10	

    # def isEqual(self, carB) :
    #     if self.color == carB.color : return True
    #     else : return False
    def __eq__(self, carB) : return self.color == carB.color  # 비교 연산자 == 중복
    def __str__(self):
        return f"color = {self.color:s}, speed = {self.speed:d}"
car2 = Car('red', 120)
car3 = Car('blue', 4)
car6 = Car('blue')

print("car2==car6 : ", car2==car6)
print("car3==car6 : ", car3==car6)

print("[car]", car3)
```

## 2.12 상속
* SW 재사용을 위한 것

```python
# 클래스의 상속
class SuperCar(Car):
    def __init__(self, color, speed=0, bTurbo=True):
        super().__init__(color, speed)      # 부모(Car)클래스의 생성자 호출
        self.bTurbo = bTurbo                # 터보모드를 위한 변수 생성 및 초기화

    def speedUp(self):
        if self.bTurbo:
            self.speed += 50
        else:
            super().speedUp()
            
s1 = SuperCar("Gold", 0, True)
s2 = SuperCar("White", 0, False)

s1.speedUp()
s2.speedUp()
print("슈퍼카1:", s1)
print("슈퍼카2:", s2)
```
